```python
trafficking = pandas.read_csv('trafficking.csv',low_memory=False)
trafficking = trafficking.drop(trafficking.columns[[0, 2]], axis=1)
trafficking = trafficking.replace(['-99','18--20', '24--26', '30--38', '21--23',
                           '27--29', '9--17','39--47', '48+', '0--8',
                           'Unknown'], ['Unknown','Adult-', 'Adult', 'Adult', 'Adult',
                                        'Adult', 'Youth', 'Adult', 'Adult+', 'Youth', 
                                            'ZZ'])
trafficking = trafficking.rename(columns={'yearOfRegistration':"Year"})
trafficking = trafficking[['Year', 'gender', 'ageBroad','citizenship','CountryOfExploitation']]
trafficking.head(5)
```

```python
traffickingCounts = trafficking.groupby(['citizenship', "Year","CountryOfExploitation"],
                 as_index=False).count().iloc[:,:-1]\
                .rename(columns={'citizenship':'properties.code','gender':'CaseCount'})
#                        .set_index(['citizenship', 'Year']) \
traffickingCounts.head()
```

```python
longcounts = traffickingCounts[['properties.code', "Year"]].drop_duplicates()
for index,row in traffickingCounts.iterrows():

    col = row[2]
    longcounts.loc[(longcounts['properties.code'] == row[0]) & (longcounts["Year"] == row[1]),
                   "Total"] = traffickingCounts.loc[
                            (traffickingCounts['properties.code'] == row[0]) &
                            (traffickingCounts["Year"] == row[1])]['CaseCount'].sum()

#     EACH COUNTRY

    longcounts.loc[(longcounts['properties.code'] == row[0]) &  (longcounts["Year"] == row[1]),
               col] = traffickingCounts.loc[
                        (traffickingCounts['properties.code'] == row[0]) &
                        (traffickingCounts["Year"] == row[1]) &
                        (traffickingCounts['CountryOfExploitation'] == col)]['CaseCount'].values[0]



longcounts.reset_index().to_json("TraffickingCounts.json", orient="records")
```

```python
 def colorPick(gdp_md_est):
        

    if gdp_md_est > 1088982000000:
        color = "#7FFF00"
    elif gdp_md_est > 367301000000:
        color = "#98FB98"
    elif gdp_md_est > 171588000000:
        color = "#87CEFA"
    elif gdp_md_est > 71926000000:
        color = "#F0E68C"
    elif gdp_md_est > 34633000000:
        color = "#F4A460"
    elif gdp_md_est > 19688000000:
        color = "#FF4500"
    elif gdp_md_est > 3165000000:
        color = "#8B0000"
    elif type(gdp_md_est) != "float64":
        color = "#FFFFFF"
    
    
    else:
        color = '#FFFFFF'

    return color


```

```python
def halfwayPoints(df, column):
    allMidPoints = []
    for _, row in df.iterrows():
        midPoints = []
        lastcoord = 0
        counter = 0
        for coord in row[column][0]:
                print(coord)
                
                if lastcoord != coord and len(coord) == 2:
                    lastcoord = coord

                    midpoint = [float(lastcoord[0])+float(coord[0]) * 0.5,
                                 float(lastcoord[1]), float(coord[1]) * 0.5 ]
            
                    midPoints.append(midpoint)
                elif  lastcoord != coord[0]:
                    lastcoord = coord[0]

                    midpoint = [float(lastcoord[0])+float(coord[0][0]) * 0.5,
                                 float(lastcoord[1]), float(coord[0][1]) * 0.5 ]
            
                    midPoints.append(midpoint)

        allMidPoints.append(midPoints)
    return allMidPoints

```

```python
# dist = math.hypot(x2 - x1, y2 - y1)
def arrowPath (df):
    path_coords = []

    for _, row in df.iterrows():
        path_coord = []
        shortest_path = 4e8
        
        if row['TraffickingStats'] != "None":
            for code in row['TraffickingStats']:
                if code != "ZZ" and code !=  "Total" and code != row['properties.code']:
                    try:
    
                        originCoords = row['geometry.coordinates']
                        destinationCoords = df[df['properties.code']==code]['geometry.coordinates']

                        for polygonA in originCoords:
                            for polygonB in destinationCoords:
    
                                while len(polygonA[0]) != 2 or len(polygonA) == 1:
                                    polygonA = polygonA[0]

                                while len(polygonB[0]) != 2 or len(polygonB) == 1:
                                    polygonB = polygonB[0]

                                for origin in polygonA:
                                    for destination in polygonB:
                                        path =  math.hypot(destination[0] - origin[0],
                                                       destination[1] - origin[1])
                                        if path < shortest_path and path != 0:
                                            shortest_path=path
                                            path_coord = [code, origin, destination, row['TraffickingStats'][code]]

                    except IndexError:
                        print("No coords for", row['properties.code'], code )
                    except KeyError:
                        print("No coords for", row['properties.code'], code )
        path_coords.append(path_coord)
        
    return(path_coords)
```

```python

globeFinal['arrowPath'] = arrowPath(globeFinal)
```

```python
def df_to_geojson(df, properties):
    geojson = {'type':'FeatureCollection', 'features':[]}
    for _, row in df.iterrows():
        feature = {'type':'Feature',
                   'properties':{},
                   'geometry':{'type': [],'coordinates':[]}}
        feature['geometry']['coordinates'] = row['geometry.coordinates']
        feature['geometry']['type'] = row['geometry.type']
        for prop in properties:
            feature['properties'][prop] = row[prop]
        geojson['features'].append(feature)
    return geojson



        
```

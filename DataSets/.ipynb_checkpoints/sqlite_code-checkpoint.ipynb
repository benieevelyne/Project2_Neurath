```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
df_gdp = pd.read_csv("Countries_GDP_Cleaned.csv",index_col=0)
df_gdp.head()
```

```python
df_mig_age_gen = pd.read_csv("MigrationCountsAge-Gender.csv", index_col=0)
df_mig_age_gen.head()
```

```python
df_mig_year = pd.read_csv("NumberPeopleMigratedFinal.csv", index_col=0)
df_mig_year.head()
```

```python
df_trafficking_counts = pd.read_csv("TraffickingCountsAge-Gender.csv", index_col=0)
df_trafficking_counts.head()
```

```python
engine = create_engine("sqlite:///db.sqlite")
```

```python
df_trafficking_counts.to_sql("trafficking_counts",engine)
```

```python
df_mig_year.to_sql("df_mig_year", engine)
```

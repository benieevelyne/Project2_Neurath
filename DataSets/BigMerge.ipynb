```python
trafficking = pandas.read_csv('trafficking.csv',low_memory=False)
# trafficking.columns = ['code', 'Year', 'CountryOfExploitation', 'TraffickedGender',
#                        'TraffickedageBroad', 'TraffickedCount'  ]
# trafficking['code'] = trafficking['code'].astype(str)
# trafficking = trafficking.sort_values(['code', 'Year'])


trafficking = trafficking.drop(trafficking.columns[[0, 2]], axis=1)
trafficking = trafficking.replace(['-99','18--20', '24--26', '30--38', '21--23',
                           '27--29', '9--17','39--47', '48+', '0--8',
                           'Unknown'], ['Unknown','Adult-', 'Adult', 'Adult', 'Adult',
                                        'Adult', 'Youth', 'Adult', 'Adult+', 'Youth', 
                                            'ZZ'])
trafficking = trafficking.rename(columns={'yearOfRegistration':"Year"})
# TraffickingCounts = 


# trafficking = trafficking.set_index('citizenship')
# trafficking.dtypes
# trafficking['Year'] = trafficking['Year'].astype(str)
trafficking = trafficking[['Year', 'gender', 'ageBroad','citizenship','CountryOfExploitation']]
trafficking.head(5)
```

```python
traffickingCounts = trafficking.groupby(['citizenship', "Year","CountryOfExploitation"],
                 as_index=False).count().iloc[:,:-1]\

                .rename(columns={'citizenship':'properties.code','gender':'CaseCount'})
#                        .set_index(['citizenship', 'Year']) \
traffickingCounts.head()
```

```python
longcounts = traffickingCounts[['properties.code', "Year"]].drop_duplicates()
for index,row in traffickingCounts.iterrows():

    col = row[2]
    
#     TOTAL
    longcounts.loc[(longcounts['properties.code'] == row[0]) & (longcounts["Year"] == row[1]),
                   "Total"] = traffickingCounts.loc[
                            (traffickingCounts['properties.code'] == row[0]) &
                            (traffickingCounts["Year"] == row[1])]['CaseCount'].sum()

#     EACH COUNTRY

    longcounts.loc[(longcounts['properties.code'] == row[0]) &  (longcounts["Year"] == row[1]),
               col] = traffickingCounts.loc[
                        (traffickingCounts['properties.code'] == row[0]) &
                        (traffickingCounts["Year"] == row[1]) &
                        (traffickingCounts['CountryOfExploitation'] == col)]['CaseCount'].values[0]



longcounts.reset_index().to_json("TraffickingCounts.json", orient="records")
```

```python
 def colorPick(gdp_md_est):
        
# THESE COLORS DONT SEEM TO MATCH...CHECK US COLORS BELOW

# divided 190 countries into 7 equal categories
     
#     (190.0, 166.375, 142.75, 119.125, 95.5, 71.875, 48.25, 24.625, 1.0)
#  GDP(ppp) categories based on 2017 numbers

    if gdp_md_est > 1088982:
        color = "CHARTREUSE"
    elif gdp_md_est > 367301:
        color = "PALEGREEN"
    elif gdp_md_est > 171588:
        color = "SKYBLUE"
    elif gdp_md_est > 71926:
        color = "PALEVIOLETRED"
    elif gdp_md_est > 34633:
        color = "SANDYBROWN"
    elif gdp_md_est > 19688:
        color = "ORANGERED"
    elif gdp_md_est > 3165:
        color = "RED"
    else:
        color = 'slategrey'

    return color


```

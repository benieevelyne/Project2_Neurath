```python
trafficking = pandas.read_csv('trafficking.csv',low_memory=False)
# trafficking.columns = ['code', 'Year', 'CountryOfExploitation', 'TraffickedGender',
#                        'TraffickedageBroad', 'TraffickedCount'  ]
# trafficking['code'] = trafficking['code'].astype(str)
# trafficking = trafficking.sort_values(['code', 'Year'])


trafficking = trafficking.drop(trafficking.columns[[0, 2]], axis=1)
trafficking = trafficking.replace(['-99','18--20', '24--26', '30--38', '21--23',
                           '27--29', '9--17','39--47', '48+', '0--8',
                           'Unknown'], ['Unknown','Adult-', 'Adult', 'Adult', 'Adult',
                                        'Adult', 'Youth', 'Adult', 'Adult+', 'Youth', 
                                            'Unknown'])
trafficking = trafficking.rename(columns={'yearOfRegistration':"Year"})
# TraffickingCounts = 


# trafficking = trafficking.set_index('citizenship')
# trafficking.dtypes
# trafficking['Year'] = trafficking['Year'].astype(str)
trafficking = trafficking[['Year', 'gender', 'ageBroad','citizenship','CountryOfExploitation']]
trafficking.head(5)
```

```python
traffickingCounts = trafficking.groupby(['citizenship', "Year"], as_index=True).sum().iloc[:,:0]
for index,row in trafficking.iterrows():
    col = str(row[4]) + str(row[1]) + str(row[2]) + 'Trafficked'

    if col not in traffickingCounts.columns:
        print(col)
        traffickingCounts[col] = 0

    traffickingCounts.loc[(row[3], row[0]), col] += 1

```

```python
traffickingCounts.head()
```

```python
Total = traffickingCounts.T.sum().reset_index().rename(columns={0:"TotalTrafficked"})\
.set_index(['citizenship', 'Year'])

TotalFemale = traffickingCounts[traffickingCounts.columns[traffickingCounts \
                                            .columns.to_series().str.contains('Female')]] \
                                            .T.sum().reset_index().rename(columns={0:"TotalFemaleTrafficked"})\
.set_index(['citizenship', 'Year'])

TotalMale = traffickingCounts[traffickingCounts.columns[traffickingCounts \
                                            .columns.to_series().str.contains('Male')]] \
                                            .T.sum().reset_index().rename(columns={0:"TotalMaleTrafficked"})\
.set_index(['citizenship', 'Year'])

TotalChild =  traffickingCounts[traffickingCounts.columns[traffickingCounts \
                                            .columns.to_series().str.contains('Child')]] \
                                            .T.sum().reset_index().rename(columns={0:"TotalChildTrafficked"})\
.set_index(['citizenship', 'Year'])


TotalAdultMinus =  traffickingCounts[traffickingCounts.columns[traffickingCounts \
                                            .columns.to_series().str.contains('Adult-')]] \
                                            .T.sum().reset_index().rename(columns={0:"TotalAdult-Trafficked"})\
.set_index(['citizenship', 'Year'])

TotalAdult =  traffickingCounts[traffickingCounts.columns[traffickingCounts \
                                            .columns.to_series().str.contains('AdultTraf')]] \
                                            .T.sum().reset_index().rename(columns={0:"TotalAdultTrafficked"})\
.set_index(['citizenship', 'Year'])

TotalAdultPlus =  traffickingCounts[traffickingCounts.columns[traffickingCounts \
                                            .columns.to_series().str.contains('Adult+')]] \
                                            .T.sum().reset_index().rename(columns={0:"TotalAdult+Trafficked"})\
.set_index(['citizenship', 'Year'])


TotalMaleChild =  traffickingCounts[traffickingCounts.columns[traffickingCounts \
                                            .columns.to_series().str.contains('MaleChild')]] \
                                            .T.sum().reset_index().rename(columns={0:"TotalMaleChildTrafficked"})\
.set_index(['citizenship', 'Year'])


TotalMaleAdultMinus =  traffickingCounts[traffickingCounts.columns[traffickingCounts \
                                            .columns.to_series().str.contains('MaleAdult-')]] \
                                            .T.sum().reset_index().rename(columns={0:"TotalMaleAdult-Trafficked"})\
.set_index(['citizenship', 'Year'])

TotalMaleAdult =  traffickingCounts[traffickingCounts.columns[traffickingCounts \
                                            .columns.to_series().str.contains('MaleAdultTraf')]] \
                                            .T.sum().reset_index().rename(columns={0:"TotalMaleAdultTrafficked"})\
.set_index(['citizenship', 'Year'])

TotalMaleAdultPlus =  traffickingCounts[traffickingCounts.columns[traffickingCounts \
                                            .columns.to_series().str.contains('MaleAdult+')]] \
                                            .T.sum().reset_index().rename(columns={0:"TotalMaleAdult+Trafficked"})\
.set_index(['citizenship', 'Year'])

TotalFemaleChild =  traffickingCounts[traffickingCounts.columns[traffickingCounts \
                                            .columns.to_series().str.contains('MaleChild')]] \
                                            .T.sum().reset_index().rename(columns={0:"TotalFemaleChildTrafficked"})\
.set_index(['citizenship', 'Year'])


TotalFemaleAdultMinus =  traffickingCounts[traffickingCounts.columns[traffickingCounts \
                                            .columns.to_series().str.contains('FemaleAdult-')]] \
                                            .T.sum().reset_index().rename(columns={0:"TotalFemaleAdult-Trafficked"})\
.set_index(['citizenship', 'Year'])

TotalFemaleAdult =  traffickingCounts[traffickingCounts.columns[traffickingCounts \
                                            .columns.to_series().str.contains('FemaleAdultTraf')]] \
                                            .T.sum().reset_index().rename(columns={0:"TotalFemaleAdultTrafficked"})\
.set_index(['citizenship', 'Year'])

TotalFemaleAdultPlus =  traffickingCounts[traffickingCounts.columns[traffickingCounts \
                                            .columns.to_series().str.contains('FemaleAdult+')]] \
                                            .T.sum().reset_index().rename(columns={0:"TotalFemaleAdult+Trafficked"})\
.set_index(['citizenship', 'Year'])


FinalTotals = pandas.concat([Total, TotalChild, TotalAdult,
                 TotalAdultMinus, TotalAdultPlus, TotalFemale, TotalMale,
                 TotalMaleChild, TotalMaleAdultMinus, TotalMaleAdult, TotalMaleAdultPlus,
                TotalFemaleChild,TotalFemaleAdultMinus, TotalFemaleAdult, TotalFemaleAdultPlus],axis=1, join='outer')
```

```python
FinalTrafficking = pandas.concat([FinalTotals, traffickingCounts], axis=1, join='outer')

FinalTrafficking = FinalTrafficking.reset_index().rename(columns={'citizenship': 'code'})
FinalTrafficking.head()
```

```python
migrationB = pandas.read_csv('NumberofPeopleMigrated_PerCountries_PerYear.csv')
migrationB = migrationB.rename(columns={'Years': 'Year', '2': "Origin", 'value':'TotalMigrants'})
migrationB = migrationB.drop('Unnamed: 0', axis=1)
MigrantNums = migrationB['TotalMigrants'].replace("..", 0)
# migrationB = migrationB.set_index(['Origin', 'Year'])


migrationB['TotalMigrants'] = migrationB['TotalMigrants'].apply(lambda x: x.replace('..','0'))
migrationB['TotalMigrants'] = migrationB['TotalMigrants'].apply(lambda x: pandas.to_numeric(x.replace(',','')))
```

```python
migrationB2 =  pandas.pivot_table(migrationB, index=['Origin', 'Year'], columns='Countries of Destination', values='TotalMigrants')
migrationB2 =  migrationB2.reset_index()
migrationB2.head()
```

```python
FinalMigration = migrationA.join(migrationB2, lsuffix='', rsuffix='_r').drop(['Origin_r', 'Year_r'], axis=1)
FinalMigration
```

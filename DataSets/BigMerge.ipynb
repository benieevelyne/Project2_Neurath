```python
 def colorPick(gdp_md_est):

# divided 190 countries into 7 equal categories
     
#     (190.0, 166.375, 142.75, 119.125, 95.5, 71.875, 48.25, 24.625, 1.0)
#  This is total gdp, in millijons

    if gdp_md_est > 500000:
        color = "CHARTREUSE"
    elif gdp_md_est > 200000:
        color = "PALEGREEN"
    elif gdp_md_est > 70000:
        color = "SKYBLUE"
    elif gdp_md_est > 35000:
        color = "PALEVIOLETRED"
    elif gdp_md_est > 15000:
        color = "SANDYBROWN"
    elif gdp_md_est > 7000:
        color = "ORANGERED"
    elif gdp_md_est > 2000:
        color = "RED"
    else:
        color = 'slategrey'

    return color


```

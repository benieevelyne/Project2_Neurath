```python
 def colorPick(gdp_md_est):
        
# THESE COLORS DONT SEEM TO MATCH...CHECK US COLORS BELOW

# divided 190 countries into 7 equal categories
     
#     (190.0, 166.375, 142.75, 119.125, 95.5, 71.875, 48.25, 24.625, 1.0)
#  GDP(ppp) categories based on 2017 numbers

    if gdp_md_est > 1088982:
        color = "CHARTREUSE"
    elif gdp_md_est > 367301:
        color = "PALEGREEN"
    elif gdp_md_est > 171588:
        color = "SKYBLUE"
    elif gdp_md_est > 71926:
        color = "PALEVIOLETRED"
    elif gdp_md_est > 34633:
        color = "SANDYBROWN"
    elif gdp_md_est > 19688:
        color = "ORANGERED"
    elif gdp_md_est > 3165:
        color = "RED"
    elif type(gdp_md_est) != "float64":
        color = "Black"
    
    
    else:
        color = 'slategrey'

    return color


```

```python
# Source: 
# https://github.com/gboeing/urban-data-science/blob/3faf7e028d48cb03ddb999c5a910213c5384e7dc/17-Leaflet-Web-Mapping/leaflet-simple-demo/pandas-to-geojson.ipynb
def df_to_geojson(df, properties):
    geojson = {'type':'FeatureCollection', 'features':[]}
    for _, row in df.iterrows():
        feature = {'type':'Feature',
                   'properties':{},
                   'geometry':{'type': [],'coordinates':[]}}
        feature['geometry']['coordinates'] = [row['geometry.coordinates']]
        feature['geometry']['type'] = row['geometry.type']
        for prop in properties:
            feature['properties'][prop] = row[prop]
        geojson['features'].append(feature)
    return geojson

cols = ['properties.code', 
#         'Year', 'GDP(ppp)', 'properties.color','TraffickDict', 
        'geometry.coordinates', 'geometry.type',
       'properties.economy', 'properties.income_grp', 'properties.name',
       'properties.sovereignt', 'properties.type']

        
```

```python
with open('../static/data/globe2.geo.json', 'w', encoding='utf-8') as f:
    json.dump(df_to_geojson(globeFinal.head(25), cols), f, ensure_ascii=False)
```

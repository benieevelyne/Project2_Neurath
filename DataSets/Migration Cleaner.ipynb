```python
import pandas as pd
import numpy as np
```

```python
# gdp = pandas.read_csv('gdp.csv')
migrationIn = pd.read_csv('UN_MigrantStockByAge.csv', skiprows=15)
```

```python
migration = pd.DataFrame(migrationIn).reset_index(drop=True)
migration.head()
```

```python
migration = migration.iloc[13:]
migration.head()
```

```python
migration = migration.drop(['Sort\norder', 'Notes', 'Type of data (a)', 'Code'],
       axis=1).set_index("Major area, region, country or area of destination")
migration.head()
```

```python
migration = migration.replace("..", 0)
migration = migration.astype('int64', errors='raise')
migration.dtypes
```

```python
Year = migration.iloc[:,0:1].squeeze().rename('Year')

Total = migration.iloc[:,17:18].squeeze().rename('Total')
TotalYouth = migration.iloc[:,1:5].sum(axis=1).rename('TotalYouth')
TotalAdult = migration.iloc[:,5:14].sum(axis=1).rename('TotalAdult')
TotalElder = migration.iloc[:,14:17].sum(axis=1).rename('TotalElder')
TotalMale = migration.iloc[:,34:35].squeeze().rename("TotalMale")
TotalFemale = migration.iloc[:,51:52].squeeze().rename("TotalFemale")

MaleYouth = migration.iloc[:,18:22].sum(axis=1).rename('MaleYouth')
MaleAdult = migration.iloc[:,22:31].sum(axis=1).rename('MaleAdult')
MaleElder = migration.iloc[:,31:34].sum(axis=1).rename('MaleElder')

FemaleYouth = migration.iloc[:,35:39].sum(axis=1).rename('FemaleYouth')
FemaleAdult = migration.iloc[:,39:48].sum(axis=1).rename('FemaleAdult')
FemaleElder = migration.iloc[:,48:51].sum(axis=1).rename('FemaleElder')

# Youth 0-19  (1-5)
# Adult 20-65 (5-14)
# Elder 65+ (14-17)
# Total (18)
```

```python
MigrationSums = pd.concat([Year, Total, TotalYouth, TotalAdult, TotalElder,
                          TotalMale, MaleYouth, MaleAdult, MaleElder,
                          TotalFemale, FemaleYouth, FemaleAdult, FemaleElder], axis=1)
MigrationSums.head()
```

```python
MigrationSums.to_csv('MigrationCountsAge-Gender.csv')
```
